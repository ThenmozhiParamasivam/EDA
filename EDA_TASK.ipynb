```python
sns.pairplot(df) 
plt.show()
```

```python
plt.figure(figsize=(10, 8))  
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df['PassengerId'], kde=True, bins=30)
plt.title('Histogram of PassengerId')
plt.xlabel('PassengerId')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df['PassengerId'])
plt.title('Boxplot of PassengerId')
plt.show()
```

```python
sns.scatterplot(x='PassengerId', y='Survived', data=df)
plt.title('PassengerId vs Survival')
plt.xlabel('PassengerId')
plt.ylabel('Survived')
plt.yticks([0, 1])
plt.show()
```
